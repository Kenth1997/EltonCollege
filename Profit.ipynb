```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Public.csv', dtype='string')
```

```python
df_for_profit = df[['for_profit', 'founded']]
df_control = df[['control', 'founded1']]
df_non_profit = df[['non_profit', 'founded2']]
```

```python
dfs = [df_for_profit, df_non_profit,df_control]
names = ['for profit', 'Non profit','control']
```

```python
counts = []
```

```python
for df, name in zip(dfs, names):
    df.columns = ['Name', 'Founded']
    df['Founded'] = pd.to_datetime(df['Founded'], yearfirst=True)
    df = df.set_index('Founded')
    df_decades = df.resample('10AS').count()
    df_decades.columns = [name]
    counts.append(df_decades)
```

```python
sns.set_style("dark")
plt.figure(figsize=(15,10))
sns.set_palette(["#090364", "#ff0000"])
sns.lineplot(data=pd.concat(counts), dashes=False)
plt.show()
```
