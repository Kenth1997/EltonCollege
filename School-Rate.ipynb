```python
%matplotlib inline
import datetime
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import random
import pylab
from matplotlib import dates
dates.YearLocator
```

```python
df = pd.read_csv('rate.csv')
df['year'] = pd.cut(df['founded_in'].astype(int), range(1700,2020,10)).apply(lambda x:x.left)
df.groupby('founded_in')['name'].count()
```

```python
plt.figure(figsize=(10,10))
sns.set_style("dark")
plt.rcParams["figure.autolayout"] = True
df.groupby('year')['name'].count().plot.bar()
pylab.xlabel('Year Founded',fontsize=25)
pylab.ylabel('No. Of Schools',fontsize=25)
pylab.title('School Rate Over Decade',fontsize=30)
```
